#Sample Functions to Plot DonkeyCar Data

```python
from numpy import arange
#from pylab import plot,show
import matplotlib.pyplot as plt
from pathlib import Path
import os
import json
import glob

TUB_DIR = '/Users/dan/mycar/data/tub-dan'

sample_size = 2000
x = arange(0,sample_size)
y = []

tubPath = Path(TUB_DIR)
json_files = glob.glob1(tubPath,"record_*.json")
sample_files = json_files[:sample_size]

for filename in sample_files:
    data = json.load(open(os.path.join(tubPath, filename)))
    y.append(data["user/angle"])
    
plt.scatter(x, y, marker=".")
plt.xlabel("File")
plt.ylabel("Angle")
plt.show()
```

```python
# Histogram
# Basic plot
import numpy as np
import matplotlib.pyplot as plt
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))
plt.bar(y_pos, height, color=(0.2, 0.4, 0.6, 0.6))
 
# Custom Axis title
plt.xlabel('X Axis Title', fontweight='bold', color = 'red', fontsize='17', horizontalalignment='center')
plt.show()
```
